### ALGERIAN FOREST FIRES DATASET
THIS INCLUDES 244 instances that regroup data of two regions of Algeria(Bejaia located in the northeast and Sidi Bel-Abbes located in northwest) from the period June 2012 to september 2012

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head()
```

```python
df.info()
```

### PERFORMING Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

Adding new column "region" as dataset is divided based on Region(bejaia and sidi-bel-abbes) (0 to 122 and 122 to last)

```python
df.loc[:122,"Region"]=0
df.loc[122:, "Region"]=1
df2 = df
```

```python
df2.info()
```

```python
df2.head()
```

```python
df2[['Region']] = df2[['Region']].astype(int)
```

```python
df2.tail()
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna().reset_index(drop=True)
```

```python
df2.isnull().sum()
```

```python
df2.iloc[[122]]
```

```python
df2 = df2.drop(122).reset_index(drop=True)
```

```python
df2.iloc[[122]]
```

```python
df2.columns
```

```python
df2.columns = df2.columns.str.strip()
df2.columns
```

```python
df2.info()
```

```python
df2[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]= df2[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df2.info()
```

```python
objects = [features for features in df2.columns if df2[features].dtypes=='O']
objects
```

```python
for f in objects:
    if f != 'Classes':
        df2[f] = df2[f].astype(float)
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2.head()
```

```python
df2.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index = False)
```

#### plot density plot for all features

```python
plt.style.use('seaborn-v0_8')
df2_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
percentage = df2_copy['Classes'].value_counts(normalize=True)*100
classlabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie chart of classes")
plt.show()
```

```python
df2_copy.corr()
```

```python
sns.heatmap(df2_copy.corr(),annot=True)
```

```python
df2['Classes'] = np.where(df2['Classes'].str.contains('not fire'),'not fire','fire')
dftemp = df2.loc[df2['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df2)
plt.ylabel("Number of fires",weight="bold")
plt.xlabel('Month',weight='bold')
plt.title("fire analysis of sidi-bel-abbes",weight='bold')
```

it is observed that August and September had the modst number of forest fires for both regions. and from the above plot of months, we can understand a few things.
